```python
epsilons = [0.001, 0.01, 0.1, 0.2, 0.5]
alphas = [0.01, 0.1, 0.2, 0.5, 0.99]

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

for (epsilon, alpha) in zip(epsilons, alphas):
    
    taxi = Taxi()
    taxi.train(epsilon=epsilon)
    xs = range(0, taxi.n_episode, 100)
    ys = taxi.log_reward[0::100]
    ax1.plot(xs, ys, label = f"Œµ = {epsilon}")

    taxi = Taxi()
    taxi.train(alpha=alpha)
    xs = range(0, taxi.n_episode, 100)
    ys = taxi.log_reward[0::100]
    ax2.plot(xs, ys, label = f"Œ± = {alpha}")

ax1.legend()
ax2.legend()
plt.show()
```

```python
epsilons = [0.001, 0.01, 0.1, 0.2, 0.5]
alphas = [0.01, 0.1, 0.2, 0.5, 0.99]

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)

for (epsilon, alpha) in zip(epsilons, alphas):
    
    taxi = Taxi()
    taxi.train(method = "SARSA", epsilon=epsilon)
    xs = range(0, taxi.n_episode, 100)
    ys = taxi.log_reward[0::100]
    ax1.plot(xs, ys, label = f"Œµ = {epsilon}")

    taxi = Taxi()
    taxi.train(method = "SARSA", alpha=alpha)
    xs = range(0, taxi.n_episode, 100)
    ys = taxi.log_reward[0::100]
    ax2.plot(xs, ys, label = f"Œ± = {alpha}")

ax1.legend()
ax2.legend()
plt.show()
```
